```python
df["Borough"] = df["Borough"].map(borough_lookup)
df
```

```python
df_borough = df.groupby(['Borough', 'Month'], as_index=False).sum(numeric_only=True)
df_borough
```

```python
london_trend = df_borough.groupby('Month').mean(numeric_only=True).reset_index()
london_trend
```

```python
merged = df_borough.merge(london_trend, on='Month', suffixes=('', '_LondonAvg'))

# Compute deviation from London average for each category
crime_categories = [col for col in df_borough.columns if col not in ['Borough', 'Month']]
for cat in crime_categories:
    merged[f'{cat}_Deviation'] = merged[cat] - merged[f'{cat}_LondonAvg']
```

```python
merged = merged.sort_values(['Borough', 'Month'])

for cat in crime_categories:
    merged[f'{cat}_Deviation_Change'] = merged.groupby('Borough')[f'{cat}_Deviation'].diff()
```

```python
borough_exceptionality = (
    merged.groupby('Borough')[[f'{cat}_Deviation_Change' for cat in crime_categories]]
    .std()
    .sort_values(by=f'{crime_categories[0]}_Deviation_Change', ascending=False)
)

borough_exceptionality
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: pick a borough
borough_to_plot = 'Hackney'

# Filter for that borough
borough_data = merged[merged['Borough'] == borough_to_plot]

# Set plot style
sns.set(style="whitegrid")
plt.figure(figsize=(12,6))

# Plot deviation over time for BURGLARY
plt.plot(borough_data['Month'], borough_data['BURGLARY_Deviation'], marker='o', label='Deviation')
plt.plot(borough_data['Month'], borough_data['BURGLARY_Deviation_Change'], marker='x', label='Month-to-month change')

plt.title(f'Time-Varying Exceptionality in BURGLARY - {borough_to_plot}')
plt.xlabel('Month')
plt.ylabel('Deviation / Change from London Avg')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
merged['BURGLARY_Deviation_Roll'] = merged.groupby('Borough')['BURGLARY_Deviation'].rolling(3, min_periods=1).mean().reset_index(0,drop=True)
```

```python
# Absolute deviation as simple quality measure
merged['BURGLARY_Quality'] = merged['BURGLARY_Deviation_Roll'].abs()
```

```python
top_boroughs_per_month = (
    merged.groupby('Month').apply(lambda x: x.nlargest(3, 'BURGLARY_Quality'))
    .reset_index(drop=True)
)
```

```python
# Example: track Hackney over time
hackney = merged[merged['Borough'] == 'Hackney']

# Identify periods of exceptionality
threshold = hackney['BURGLARY_Quality'].quantile(0.75)  # top 25% = exceptional
hackney['Is_Exceptional'] = hackney['BURGLARY_Quality'] > threshold

# Detect start and stop periods
hackney['Start_of_exception'] = hackney['Is_Exceptional'] & (~hackney['Is_Exceptional'].shift(1).fillna(False))
hackney['End_of_exception'] = (~hackney['Is_Exceptional']) & (hackney['Is_Exceptional'].shift(1).fillna(False))
```
